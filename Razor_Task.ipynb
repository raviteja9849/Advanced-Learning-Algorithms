```python
 # excluding  the header row


df = df.iloc[1:]
```

```python
# By executing this code, we will delete the columns 'Date Added', 'sellerlink', 'sellerlink-url', and 'sellerstorefront-url'.

# We have delete these rows because those columns are not much helpful in our data visualization and analysis
# and also to optimize the data for better results

# Finally , it will drop the specified columns from the DataFrame df and reploace  the result to assign the result
# again to df column

df = df.drop(['Date Added' , 'sellerlink' , 'sellerlink-url' , 'sellerstorefront-url'] , axis=1)


```

```python
# Perform data cleaning and parsing on the column category


df['category'] = df['category'].str.strip()


# Fill null values in the 'category' column with 'other' as category

df['category'] = df['category'].fillna('other')


print(df['category'])
```

```python
# Extracting Total product count from the column sellerproductcount and storing those values in
# The existing  column called with replacement. Also fiiling the null values with 0.



df['sellerproductcount'] = df['sellerproductcount'].str.extract(r'(\d+)(?=\sresults)').astype(float).fillna(0).astype(int)
```

```python
import pandas as pd
import re

# Assuming your DataFrame is named df
df['Total Positive Seller Ratings Percentage'] = df['sellerratings'].str.extract(r'^(\d+)%')
df['Total Positive Seller Ratings'] = df['sellerratings'].str.extract(r'\((\d+) ratings\)$')

# Convert the extracted columns to numeric values

df['Total Positive Seller Ratings Percentage'] = pd.to_numeric(df['Total Positive Seller Ratings Percentage'], errors='coerce')
df['Total Positive Seller Ratings'] = pd.to_numeric(df['Total Positive Seller Ratings'], errors='coerce')

# Fill any NaN values with 0

df['Total Positive Seller Ratings Percentage'].fillna(0, inplace=True)
df['Total Positive Seller Ratings'].fillna(0, inplace=True)


# deleting the column sellerratings as we have extracted the necessary information

df.drop('sellerratings', axis=1, inplace=True)
```

```python
# importing the library called re for regular expressions.

import re


# Extract telephone numbers using regex
telephone_numbers = df['sellerdetails'].str.extract(r'(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})')

# Rename the extracted column
telephone_numbers.columns = ['Phone Number']

# Attach the extracted telephone numbers to the original DataFrame
df['Phone Number'] = telephone_numbers['Phone Number']


# In this code, we use the str.extract() function to apply a regex pattern to the
# 'Seller Details' column and extract the telephone numbers. The regex pattern \d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}
# matches phone number formats like XXX-XXX-XXXX, XXX.XXX.XXXX, or XXX XXX XXXX. The extracted telephone numbers
# are stored in a new DataFrame called telephone_numbers, and the column is renamed to 'Phone Number'.
# Finally, we display the extracted telephone numbers.
```

```python
# Extract email IDs using regular expression pattern



# Adding another new column called 'Email ID' to the DataFrame df.
# We again use the str.extract() method to extract email IDs from the 'sellerdetails' column using a regular expression pattern.
# The pattern r'Email:\s*([\w\.-]+@[\w\.-]+)' captures email addresses in the format of name@domain
# after the keyword 'Email:' and any whitespace characters (\s*).


df['Email ID'] = df['sellerdetails'].str.extract(r'Email:\s*([\w\.-]+@[\w\.-]+)')

```

```python
# Create a boolean mask for non-null values in the 'Email ID' column

mask = df['Email ID'].notnull()

# Display the non-null email IDs

non_null_emails = df.loc[mask, 'Email ID']
print(non_null_emails)
```

```python
# Create a boolean mask for non-null values in the 'Phone Number' column
mask = df['Phone Number'].notnull()

# Display the non-null phone numbers
non_null_phone_numbers = df.loc[mask, 'Phone Number']
print(non_null_phone_numbers)
```

```python
# deleting the column sellerdetails as we have extracted the necessary information

df.drop('sellerdetails', axis=1, inplace=True)
```

```python
# In this code, we access the 'Seller Business Name' column of the DataFrame df.
# We split each value in the column based on the colon (':') using the str.split(':') method.
# Then, we extract the last part of the split using str[-1] to keep the original name after the colon.
# Finally, we remove any leading or trailing whitespace using str.strip(). The 'Seller Business Name'
# column will now contain only the cleaned business names without the part before the colon and the colon symbol itself.


df['seller business name'] = df['seller business name'].str.split(':').str[-1].str.strip()
```

```python
# printing the seller business name column after removing the unnecessary part

print(df['seller business name'])
```

```python
# Create a new column 'Business Location' and extract country codes

df['Business Location'] = df['businessaddress'].str.extract(r'(US|DE|CN)', flags=re.IGNORECASE)



# Replace NaN values in 'Business Location' with an empty string

df['Business Location'] = df['Business Location'].fillna('')



# Map the country codes to their respective locations

country_mapping = {'US': 'United States', 'DE': 'Germany', 'CN': 'China'}
df['Business Location'] = df['Business Location'].map(country_mapping)
```

```python
# This code filters the DataFrame based on the condition where the value in the 'Business Location'
# column is not equal to 'China'. It creates a new DataFrame with only the rows where the business location is not China.


df = df[df['Business Location'] != 'China']
```

```python
# deleting the column businessaddress as we have extracted the necessary information

df.drop('businessaddress', axis=1, inplace=True)
```

```python
# tO IDENTIFY THE TOP 30 BEST SELLERS BASED ON NORMALIZATION



# [1] : Define the criteria: Determine the factors that contribute to a seller's performance and desirability.

# [2] Normalize the data: Normalize the relevant columns to bring them to a common scale.
# This ensures that each criterion has equal weightage in the final evaluation.


# [3] Assign weights: Assign weights to each criterion based on their relative importance.
# This allows US to emphasize certain factors over others.WE ARE  MakING sure the weights add up to 1.

# [4] Calculate scores: Multiply the normalized values of each criterion with their corresponding
# weights and sum them up to calculate a composite score for each seller.

# [5] Sort and select top sellers: Sort the sellers based on their composite scores
# in descending order and select the top 30 sellers with the highest scores.



# Define the criteria and their weights
weights = {
    'Total Positive Seller Ratings': 0.3,
    'sellerproductcount': 0.2,
    'Hero Product 1 #ratings': 0.2,
    'Hero Product 2 #ratings': 0.2,
    'Max % of negative seller ratings - last 12 months': 0.1
}

# Normalize the data
normalized_data = df.copy()  # Assuming df is your DataFrame with relevant columns
for col in weights.keys():
    normalized_data[col] = (normalized_data[col] - normalized_data[col].min()) / (normalized_data[col].max() - normalized_data[col].min())

# Calculate the composite score
normalized_data['Composite Score'] = normalized_data.apply(lambda row: sum(row[col] * weight for col, weight in weights.items()), axis=1)

# Select the top 30 sellers
top_30_sellers = normalized_data.nlargest(30, 'Composite Score')


print(top_30_sellers)
```
